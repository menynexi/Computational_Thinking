```python
#step one import libraries
#facilitates data science
import pandas as pd

#facilitates math
import numpy as np

# New Library!!
import matplotlib.pyplot as plt

#this will display graphs inline (in this notebook)
%matplotlib inline

#specify how large to show figures (length, width)
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
#step two read the dataset and assign it to a variable 
#I used an absolute path to make it easier for the library to detet where the dataset was but is optional 
#note please make sure the directory is the same as mine when testing or edit to match the directory in which the dataset is located
pokemon = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/pokemon_viz.csv")
df = pd.DataFrame(data = pokemon)
```

Here are where I noticed Some minor problems with the dataset, pokemon by generation are classified by its original generation. For example VenusaurMega Venusaur is part of the MEGA evolutions which came from Generation 6 but it is classified as Generation 1 because Venusaur first apperaed in Gen 1. So the question now becomes *which of Generation is the strongest based on their original pokemon.* and according to the dataset being used

### QUESTION 1: How many Pokemon are in each Generation ?

```python
pokemon_gen = pokemon["Generation"].value_counts()

df['Generation'].value_counts(ascending=True).plot(kind= 'barh', width = 0.75);
plt.ylabel("Generation", labelpad=20, fontsize = 14)
plt.xlabel("# of pokemon", labelpad=20, fontsize = 14)
plt.title("Number of Pokémon per Generation", fontsize = 18);
```

```python
print(pokemon_gen)
```

### QUESTION 2: Which generation has the over all higher HP, Attack, Defense, Special Attack, Special Defence, and speed?
(this can be considered multiple sub questions)


###Which generation has a higher average HP?

```python
gen =  pokemon.groupby('Generation')['HP'].mean()
plt.title("Average Pokémon HP per Generation", fontsize = 18);
plt.xlabel("Generation", labelpad=20, fontsize = 14)
plt.ylabel("HP", labelpad=20, fontsize = 14)
gen.plot()
```

Answer : Generation 4 has the highest HP average out off all the Generations

###Which generation has a higher average Attack?

```python
gen =  pokemon.groupby('Generation')['Attack'].mean()
plt.title("Average Pokémon Attack per Generation", fontsize = 18);
plt.xlabel("Generation", labelpad=20, fontsize = 14)
plt.ylabel("Attack", labelpad=20, fontsize = 14)
gen.plot()
```

###QUESTION 4: If every generation all had the same number of pokemon which generation would be the strongest overall based on total stats?

First I will divide the dataset into generations so we can isolate them into their own list. The reason why we will isolate them is to get the strongest Total stats from every pokemon

```python
# we issolate them by ranging them from were every generation starts and ends
gen1pokemon = pokemon[:166]
gen2pokemon = pokemon[166:272]
gen3pokemon = pokemon[272:432]
gen4pokemon = pokemon[432:553]
gen5pokemon = pokemon[553:718]
gen6pokemon = pokemon[718::]
```

Then I will Issolate them even further and get the first 82 pokemon who have the highest over all total stats. The reason why I chose 82 was because that is the smallest number of pokemon in a generation.

```python
gen1 = gen1pokemon.sort_values(by = "Total", ascending = False)[0:82]
gen2 = gen2pokemon.sort_values(by = "Total", ascending = False)[0:82]
gen3 = gen3pokemon.sort_values(by = "Total", ascending = False)[0:82]
gen4 = gen4pokemon.sort_values(by = "Total", ascending = False)[0:82]
gen5 = gen5pokemon.sort_values(by = "Total", ascending = False)[0:82]
gen6 = gen6pokemon.sort_values(by = "Total", ascending = False)[0:82]
```

Then I will calculate the over all average of each generation beased on their total stats and organize them into a single list so that I can effectively plot them in a single figure.

```python
gen1total = gen1["Total"].mean()
gen2total = gen2["Total"].mean()
gen3total = gen3["Total"].mean()
gen4total = gen4["Total"].mean()
gen5total = gen5["Total"].mean()
gen6total = gen6["Total"].mean()

gennumber= [1,2,3,4,5,6]
genTotal = [gen1total,gen2total,gen3total,gen4total,gen5total,gen6total]

plt.bar(gennumber, genTotal)

plt.title("Average Pokémon Total stats per Generation", fontsize = 18);
plt.xlabel("Generation", labelpad=20, fontsize = 14)
plt.ylabel("Total", labelpad=20, fontsize = 14)
```

Answer: As you can see isolating each generation with the same number of pokemon did make a diffrence over all in which generation is the strongest. If every generation all had the same number of pokemon; Generation 3 would have the highest total stats over all.
